```python
print("====soal no 5=====")
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ekskul = ['Paskibra', 'PMR', 'PRAMUKA', 'FUTSAL', 'VOLLY', 'KESENIAN', 'ROHIS' ]
mengikuti = [5, 18, 30, 7, 16, 19, 12]
c = 'black', 'pink', 'blue', 'aqua', 'green'
plt.title("kelas 10")
plt.bar(ekskul, mengikuti, color = c)
plt.show()

print("")
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
plt.title("usia")
ekskul = ['Paskibra', 'PMR', 'PRAMUKA', 'FUTSAL', 'VOLLY', 'KESENIAN', 'ROHIS' ]
mengikuti = [5, 18, 30, 7, 16, 19, 12]
plt.title("kelas 10")
ax.pie(mengikuti,labels=ekskul,autopct='%1.2f%%')
plt.show()
```
